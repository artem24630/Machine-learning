```
import random
import matplotlib.pyplot as plt
import numpy as np

kol =20

x = [random.randint(0,100) for i in range(kol)]
y = [random.randint(0,100) for i in range(kol)]

fig = plt.figure()

for i in range(kol):
  plt.scatter(x[i],y[i])



massK = []
massB = []
for i in range(kol):
  for j in range(i+1,kol):
    if x[i] != x[j]:
      k = (y[i]-y[j])/(x[i]-x[j])#Соединяем каждую точку с каждой и находим k и b
      b = y[i] - k*x[i]
    else:
      continue
    massK+=[k]
    massB+=[b]
    

k_fin = np.array(massK).mean() #Ищем среднее k, b 
b_fin = np.array(massB).mean()

x_line = [1,100]
y_line = [k_fin+b_fin,100*k_fin+b_fin]

plt.plot(x_line,y_line)
```
